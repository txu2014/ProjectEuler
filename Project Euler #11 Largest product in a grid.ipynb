```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
S=[0]*23
for i in xrange(20):
    S[i] = map(int, raw_input().split(' '))
    S[i].extend([0,0,0])
for i in xrange(20,23):
    S[i] = [0]*23

from functools import reduce
import operator
max_prod = 0
for i in xrange(20):
    for j in xrange(20):
        #print(str(i)+","+str(j))
        prod_h = reduce(operator.mul, [S[i][j],S[i][j+1],S[i][j+2],S[i][j+3]])
        prod_v = reduce(operator.mul, [S[i][j],S[i+1][j],S[i+2][j],S[i+3][j]])
        prod_d = reduce(operator.mul, [S[i][j],S[i+1][j+1],S[i+2][j+2],S[i+3][j+3]])
        prod_rd = reduce(operator.mul, [S[i][j+3],S[i+1][j+2],S[i+2][j+1],S[i+3][j]])
        max_prod = max(prod_rd,prod_h,prod_v,prod_d,max_prod)
print(max_prod)
```
